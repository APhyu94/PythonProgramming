Ex2 : Convert Currency.
Write a program that asks the user for the amount in usd and converts it to British pounds. The exchange rate is from usd to BP is given as 0.82

Ex 4: Daily Maths
You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.




```python

Apple_price = int(input("The Price per one apple:"))
Apple_quantity = int(input("The Quanity of Apple:"))

Orange_price = int(input("The Price per one orange:"))
Orange_quantity = int(input("The Quantity of Oranges:"))

Mango_price = int(input("The Price per one mangoes:"))
Mango_quantity = int(input("The Quanitiy of Mangoes:"))

Apple_total = Apple_quantity * Apple_price
Orange_total = Orange_quantity * Orange_price
Mango_total = Mango_quantity * Mango_price

total_for_all = Apple_total + Orange_total + Mango_total

print("Total cost for apples: {} MMK".format(Apple_total))
print("Total cost for oranges: {} MMK".format(Orange_total))
print("Total cost for manages: {} MMK".format(Mango_total))
print("Total cost for all : {} MMK".format(total_for_all))



```
