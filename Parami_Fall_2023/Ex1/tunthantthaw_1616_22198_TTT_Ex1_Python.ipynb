# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python

AP = int(input("Price of the apple = "))
AQ = int(input("Quantity of the apple = "))
AF = str(AP * AQ)

OP = int(input("Price of the orange = "))
OQ = int(input("Quantity of the orange = "))
OF = str(OP*OQ)

MP = int(input("Price of the mango = "))
MQ = int(input("Quantity of the mango = "))
MF = str(MP*MQ)

total_amount = int(AF) + int(OF) + int(MF)
print("Here is your total amount! {}".format(total_amount))
```
