# Ex 3: Temperature Converter

Write a program that convert temperature in Feranhite to celsius.

```python
f_temp = input  ("Enter the temperature in Fahrenheit: ")
c_temp = int(f_temp) - 32 * 5/9
print("{} Fahrenheit is equivalent to {} Celsius.".format(f_temp, c_temp))
```

# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
q_apple = input  ("Enter the quantity for apple: ")
q_orange = input  ("Enter the quantity for orange: ")
q_mango = input  ("Enter the quantity for mango: ")

p_apple = input  ("Enter the price for apple: ")
p_orange = input  ("Enter the price for orange: ")
p_mango = input  ("Enter the price for mango: ")

total_cost = int(q_apple)*int(p_apple) + int(q_orange)*int(p_orange) + int(q_mango)*int(p_mango)

print("Total cost is equivalent to {}$.".format(total_cost))
```
